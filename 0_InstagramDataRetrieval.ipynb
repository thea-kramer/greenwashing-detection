```python
import pandas as pd
import numpy as np
```

```python
from pytangle.api import API
api = API()

# get all posts from a list from Jan until June 2020
list_ids = [1739403] # ids of the lists of interest
post_list = []
for year in range(2011,2023):
    print("collecting year",year)
    year_list = []
    for n, a_post in enumerate(api.posts(listIds=list_ids,
                                        count=-1,
                                        batchSize=100,
                                        sortBy='date',
                                        startDate= str(year)+'-01-01',
                                        endDate=str(year + 1)+'-01-01',
                                        #startDate= str(year)+'-01-01',
                                        timeframe=None,
                                        includeHistory=True
                                        )):
        # do something with the post
        year_list.append(a_post)
    data = pd.DataFrame.from_records(year_list)
    data.to_csv("results-" + str(year) + ".csv", header=True, index=False, columns=list(data.axes[1]))
data
```
