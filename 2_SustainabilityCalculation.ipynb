```python
import pandas as pd
import numpy as np
```

### Preprocessing

Change the folder either to `instagram_data/` or to `twitter_data/`

```python
folder  = 'twitter_data/' #twitter_data/ or instagram_data/
```

```python
file = folder + 'posts_restructured.csv'
data = pd.read_csv(file)
data
```

```python
#sheet: "sustainab|sustain|climate|green|environment|ecofriendly|eco-friendly|solar|co2|biodegra|renewable|nature|natural |decarbon|recycl|pollut|waste|ecolog|emission|nachhaltig|klima|umwelt|ökolog|erneuerbar|global warming|wind energy|energy efficient"

word_list_en = ['sustainab','climate','green','environment','ecofriendly','zero emission','zero-emission','clean energy','cleanenergy','ecolog','eco-friendly','biodegra','wind power','windpower','renewable','netzero','greenbuild','decarbon','recycl', 'pollut','zero waste','wind energy','co2 neutral','co2-neutral','solar', 'emissions','global warming','fairtrade','fairtrade']
word_list_de = ['nachhaltig', 'klima','umwelt','ökolog','erneuerbar','bio ','natur ','windkraft','energiespeicher','wärmepump']
word_list = word_list_en+word_list_de

word_list_en_recall = word_list_en + ['co2', 'reimaginecarbon','all-electric', 'emissionfree', 'emission free', 'planet','emobility','natur','waste']
word_list_de_recall = word_list_de + ['biologisch','grün']
word_list_high_recall = word_list_en_recall + word_list_de_recall

word_list_en_precision = [ elem for elem in word_list_en if not elem in ['green', 'climate', 'environment']]  
word_list_de_precision = [ elem for elem in word_list_de if not elem in ['klima','natur ']]  
word_list_high_precision = word_list_de_precision + word_list_en_precision
```

```python
all = data.shape[0]
print('A list of words and in how many posts they appear (percentage):')
for word in word_list:
    print(word, round(data.loc[data['post_text'].str.contains(word, na=False, case=False,regex=True)].shape[0]*100.0/all,2), "%")
```

```python
regex_query = ('|').join(word_list)
regex_query_recall = ('|').join(word_list_high_recall)
regex_query_precision = ('|').join(word_list_high_precision)
data['green'] = data['post_text'].str.contains(regex_query, na=False, case=False,regex=True)
data['green_recall'] = data['post_text'].str.contains(regex_query_recall, na=False, case=False,regex=True)
data['green_precision'] = data['post_text'].str.contains(regex_query_precision, na=False, case=False,regex=True)
#optionally, replace data.green with data.green_recall for high recall or data.green_precision for high precision
green = data.loc[data.green]
```

```python
data.to_csv(folder + 'posts_processed.csv', header=True, index=False, columns=list(data.axes[1]), lineterminator='\r\n')
green.to_csv(folder + 'green_posts_processed.csv', header=True, index=False, columns=list(green.axes[1]), lineterminator='\r\n')
```
