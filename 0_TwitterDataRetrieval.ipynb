### JSON to CSV Transformation

```python
import os, json, glob
temp = pd.DataFrame()

users = False #set to true if you need users instead of tweets

path_to_json = 'user_JSON_files/' if users else 'JSON_files/'

json_pattern = os.path.join(path_to_json,'*.json')
file_list = glob.glob(json_pattern)

temp = pd.DataFrame()
for file in file_list:
    with open(file, encoding="utf8") as f:
        file_read = f.read()
        js = json.loads(file_read)
    data = pd.DataFrame.from_records(js["users"])
    #data = pd.read_json(file, lines=True)
    temp = temp.append(data, ignore_index = True)
temp
```

```python
from datetime import datetime
dateTimeObj = str(datetime.now()).replace(" ","-").replace(":","-").replace(".","-")
temp.to_csv("twitter_users_"+dateTimeObj+".csv", header=True, index=False, columns=list(temp.axes[1]))
```

### Joining and extracting data

```python
# Joining Tweets with User Information (both from Twitter API) and Company Information (own research)
data = pd.read_csv('twitter_data/twitter_data_07-12-22.csv')
users = pd.read_csv('twitter_data/twitter_users_2022-12-07-13-17-48-602427.csv')
dax40 = pd.read_csv('twitter_data/DAX40_21-12-22.csv')
dax40 = dax40[:40]
dax40
```

```python
data.info()
```

```python
#preparing for merge
users = users.rename(columns={"id": "author_id"})
user_names = users[["author_id","username"]].drop_duplicates()
dax_attributes = dax40.rename(columns={"Twitter Username": "username"})[["username", "Branche", "Branche-6", "Sektor", "B2B", "B2C"]]

#merging
user_names = user_names.merge(dax_attributes, how="left", on="username")
data_2 = data.merge(user_names, how="left", on="author_id")

#display
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
data_2
```

```python
#more intuitive naming
data = data_2.rename(columns={"username": "author", "text": "Tweets"})
```

##### Extract public metrics

```python
import ast
df = data["public_metrics"].astype('str')
df = df.apply(lambda x: ast.literal_eval(x))
df = df.apply(pd.Series)
data[['likes','retweets','replies','quotes']] = df[['like_count','retweet_count','reply_count','quote_count']]

data
```

##### Extract referenced tweets

```python
def evaluate_referenced_tweets(referenced_tweets):
    ret = referenced_tweets
    if(referenced_tweets != 'nan'):
        ret = {'quote_ids':[], 'reply_ids':[], 'retweet_ids':[]}
        for tweet in referenced_tweets:
            if tweet['type'] == 'quoted':
                ret['quote_ids'].append(tweet['id'])
            elif tweet['type'] == 'replied_to':
                ret['reply_ids'].append(tweet['id'])
            elif tweet['type'] == 'retweeted':
                ret['retweet_ids'].append(tweet['id'])
            else:
                print('another type found: '+ tweet['type'])
    return ret

df = data["referenced_tweets"].astype('str')
df = df.apply(lambda x: ast.literal_eval(x) if x != 'nan' else x)
df = df.apply(lambda x: evaluate_referenced_tweets(x))
df = df.apply(pd.Series) 

data[['quote_ids','reply_ids','retweet_ids']] = df[['quote_ids','reply_ids','retweet_ids']]
data
```

##### Extract attachments

```python
df = data["attachments"].astype('str')
df = df.apply(lambda x: ast.literal_eval(x) if x != 'nan' else x)
df = df.apply(pd.Series)
data[['media_keys','poll_ids']] = df[['media_keys','poll_ids']]
data
```

##### Extraxt entities

```python
df = data["entities"].astype('str')
df = df.apply(lambda x: ast.literal_eval(x) if x != 'nan' else x)
df = df.apply(pd.Series)
data[['mentions','annotations','urls','hashtags','cashtags']] = df[['mentions','annotations','urls','hashtags','cashtags']]
data
```

```python
#extract mentions and hashtags
df = data["mentions"].astype('str')
df = df.apply(lambda x: [i['username'] for i in ast.literal_eval(x) if 'username' in i] if x!='nan' else x)
data['mentioned_usernames'] = df

df = data["hashtags"].astype('str')
df = df.apply(lambda x: [i['tag'] for i in ast.literal_eval(x) if 'tag' in i] if x!='nan' else x)
data['mentioned_hashtags'] = df

data
```

```python
#extract annotations
def evaluate_annotations(referenced_tweets):
    ret = {'types':[], 'entities':[]}
    for tweet in referenced_tweets:
        #print(tweet)
        if 'normalized_text' in tweet:
            ret['entities'].append(tweet['normalized_text'])
        else:
            ret['entities'].append("")
        ret['types'].append(tweet['type'])
        
    return ret

df = data["annotations"].astype('str')
df = df.apply(lambda x: ast.literal_eval(x) if x != 'nan' else x)
df = df.apply(lambda x: evaluate_annotations(x) if x!='nan' else x)
df = df.apply(pd.Series) 

data[['entity_types','entity_names']] = df[['types','entities']]
data
```

```python
#save to csv file
data.to_csv('twitter_data/tweets_extended_fields.csv', header=True, index=False, columns=list(data.axes[1]))
```
