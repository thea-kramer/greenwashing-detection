```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

data = pd.read_csv('instagram_data/Instagram_07_12_22.csv')
dax40 = pd.read_csv('DAX40_21_12_22.csv')
dax40 = dax40[:40]

data['year'] = pd.to_datetime(data['date']).dt.year
data
```

```python
import ast
df = data["account"].astype('str')
df = df.apply(lambda x: ast.literal_eval(x))
df = df.apply(pd.Series)
data['author'] = df['name']
data['username'] = df['handle']

df = data["statistics"].astype('str')
df = df.apply(lambda x: ast.literal_eval(x))
df = df.apply(pd.Series)
data['actual'] = df['actual']
df = data["actual"].astype('str')
df = df.apply(lambda x: ast.literal_eval(x))
df = df.apply(pd.Series)
data['likes'] = df['favoriteCount']
data['comments'] = df['commentCount']
data
```

```python
#remove other languages (3008 from 66949 posts were non English and non German)
data = data[(data['languageCode'] == 'en') | (data['languageCode']=='de')]
data
```

```python
dax_attributes = dax40.rename(columns={"instagram": "username"})[["username", "Branche", "Branche-6", "Sektor", "B2B", "B2C"]]
data = data.merge(dax_attributes, how="left", on="username")
```

```python
data = data.rename(columns = {"description":"post_text"})
data.info()
```

```python
data.to_csv('instagram_data/posts_restructured.csv', header=True, index=False, columns=list(data.axes[1]), lineterminator='\r\n')
```
